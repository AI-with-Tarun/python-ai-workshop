## Immutability- String

Notice that in the prior examples, we were not changing the original string with any of the operations we ran on it.

**Every string operation is defined to produce a new string as its result, because strings are immutable** in Pythonthey cannot be changed in-place after they are created. For example, you cant change a string by assigning to one of its positions, but you can always build a new one and assign it to the same name.

> Because Python cleans up old objects as you go

## args and kwargs

```
# args list_of_marks = [], list_of_product = []
# kwargs card = {key:value}, prod = {key:value}

syntax:
# args - mainly used when you have list
def function_name(*args): # consider all the elements the list as the parameter
  inside this function the args will act as tuple or list
  pass
# kwargs - mainly used when you have dictionary
def function_name(**kwargs):
  inside this function the kwargs will act as dictionary temperature = 0 {"temperature":0}
  pass
  logic
```

````python
def calculate_total_marks_average(*args):
  print(type(args)) # tuple
  total_marks = sum(args)
  average_marks = total_marks/len(args)
  return total_marks,average_marks
````

````python
calculate_total_marks_average(80,90,89,76,45,60,76,89,100)
"""
if you have multiple values to pack inside the given function, we are defining it as *
when the input is list or tuple
"""
````

## Exception: Try and Except

````python
try:
    price = int(input("Enter product price: ")) # string or float
    extra = price + 300
    print(f"Total price: {extra}")
except ValueError:
    print("Please enter a valid price!")
except:
    print("Error")
````

````python
try:
    num = {"a":10,"b":20}
    result = 10 / num['c'] # KeyError
    print("Result:", result)
except ValueError:
    print("Error: Please enter a valid number.")
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except:
    print("Invalid choice")
finally:
    print("This will run no matter what happens.")
````

## File Manipulation

Modes - in file

- w
- r
- a

````python
employee_data = "Name: Rahul Jain\nEmployee ID: EMP12345\nDepartment: IT\nSalary: 94000"
file_path = "employee_record.txt"
````

```
Syntax:

with open(file_path,mode) as any_var_name:
   any_var_name.func()
   # w and a - write
   # r - read or readlines (new lines i.e., your data is in paragraphs \n)
```

````python
with open(file_path,"w") as file: # logs #langfuse, pheonix, open
  file.write(employee_data) #str

print("Data saved")
````

````python
with open(file_path,"r") as f:
  response = f.read() # return string that is saved inside the file.
  print(response)

f.close()
````
